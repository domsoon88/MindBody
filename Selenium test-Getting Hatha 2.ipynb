```python
import time
import re
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.wait import WebDriverWait   

def init_driver(path=r'C:\xxxxxx\chromedriver.exe'):
    driver = webdriver.Chrome(path)
    return driver
```

```python
# Enter class name
yoga_class = ["Hatha2", "Vinyasa2","UpsideDown","Core","YogaWheel1"]
```

```python
start = time.time()
bbb_url = "https://clients.mindbodyonline.com/classic/mainclass?studioid=81"
driver = init_driver()
driver.get(bbb_url)
time.sleep(2)

class_list = []

# Cycle the days
for days in range(10): # no of days to go forward
    eleDate = driver.find_element_by_class_name('date-arrow-r')
    eleDate.click()


# # Select location
#     time.sleep(1)
#     eleLocation = driver.find_element_by_class_name('selectBox-arrow')
#     eleLocation.click()
#     eleLocationDropDown = driver.find_element_by_class_name('selectBox-dropdown-menu')
# #     print(type(eleLocationDropDown))
# #     print(eleLocationDropDown)
#     time.sleep(1)
#     eleAST = eleLocationDropDown.find_element_by_css_selector("[title*='Yoga - Asia Standard Tower']")
# #     print(type(eleAST))
#     eleAST.click()
#     time.sleep(2)


# select class required
    td_list = WebDriverWait(driver, 10).until(lambda driver: driver.find_elements_by_css_selector("#classSchedule-mainTable tr td"))
#     print(len(td_list))
    for items in td_list:
        class_list.append(items.text)
driver.close()
end = time.time()
print(end-start)
```

```python
# cleaning the extraspaces
cleanlist = [re.sub(' +', '', i) for i in class_list]
```

```python
index_pos = [] 
for i, v in enumerate(cleanlist): # returns index position of yoga classes
    if v in yoga_class:
        index_pos.append(i)
    if "2018" in v:
        index_pos.append(v)
```

```python
req_classes = []

for i in index_pos:
    if "2018" in str(i):
        req_classes.append((i,'','',''))
    else:        
        req_classes.extend([(cleanlist[i-2],cleanlist[i],cleanlist[i+1],cleanlist[i+3])])
```

```python
df = pd.DataFrame(req_classes, columns = ["Time","Class","Teacher","Studio"])
df.set_index('Time')
```

```python
df.to_csv('yogaclasses.csv', encoding='utf-8')
```
